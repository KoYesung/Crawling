```python
import chromedriver_autoinstaller
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from urllib.request import Request, urlopen
import os
import requests
```

```python
image_urls = []
category = ['New Arrival', 'Allitems-new', 'Allitems-small', 'Allitems-medium', 'Allitems-large']
for i in range(len(category)):
    save_folder = category[i]
    if not os.path.exists(save_folder):
        os.makedirs(save_folder)
    
    driver = webdriver.Chrome()
    driver.implicitly_wait(3)
    url = f'https://minitmute.com/product/list.html?cate_no={i+42}'
    driver.get(url)
    time.sleep(1)
    
    
    image_area_xpath = '/html/body/div[4]/div/div/div[1]'
    image_area = driver.find_element(By.XPATH, image_area_xpath)
    image_elements = image_area.find_elements(By.TAG_NAME, 'img')
    
    for image_element in image_elements:
        image_url = image_element.get_attribute('src')
        image_urls.append(image_url)
    print(image_urls)
    
    for image_url in image_urls:
        response = requests.get(image_url)
        filename = os.path.join(save_folder, image_url.split("/")[-1])
        with open(filename, "wb") as f:
            f.write(response.content)
            print(f"이미지 저장: {filename}")
```
