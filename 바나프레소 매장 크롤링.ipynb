### 다음 페이지로 이동하기

```python
# page_1_path = '/html/body/div/div/div/div/article/div/section/div/div[1]/div[3]/ul/li[1]/a'
# page_2_xpath = '/html/body/div/div/div/div/article/div/section/div/div[1]/div[3]/ul/li[2]/a'
# page_3_xpath = '/html/body/div/div/div/div/article/div/section/div/div[1]/div[3]/ul/li[3]/a'
# page_4_xpath = '/html/body/div/div/div/div/article/div/section/div/div[1]/div[3]/ul/li[4]/a'
# 6페이지로 넘어가기 위한 화살표 버튼 = /html/body/div/div/div/div/article/div/section/div/div[1]/div[3]/span[2]/a
# 11페이지로 넘어가기 위한 화살표 버튼 = /html/body/div/div/div/div/article/div/section/div/div[1]/div[3]/span[2]/a
# page_6-_path = '/html/body/div/div/div/div/article/div/section/div/div[1]/div[3]/ul/li[1]/a'
stores = []
for i in range(1, 6):
    if i > 5:
        page_xpath = '/html/body/div/div/div/div/article/div/section/div/div[1]/div[3]/span[2]/a'
    else:
        page_xpath = f'/html/body/div/div/div/div/article/div/section/div/div[1]/div[3]/ul/li[{i}]/a'
        
    driver.find_element(By.XPATH, page_xpath).click()
    time.sleep(3)
    
    soup = BeautifulSoup(driver.page_source)

    store_area = soup.findAll('span',{'class','store_name_map'})

    for store in store_area:
        i_tag = store.find('i')
        span_tag = store.find('span')
        if i_tag is not None and span_tag is not None:
            store_name = i_tag.text
            store_address = span_tag.text
            store_info = {'매장명': store_name, '매장주소': store_address}
            stores.append(store_info)
```

```python
store_info_list = []
for i in range(1, 12):
    # 숫자버튼
    button_xpath = f'//li/a[contains(text(), "{i}")]'
    button = driver.find_element(By.XPATH, button_xpath)
    button.click()

    # 가게 정보 파싱
    store_elements = driver.find_elements(By.CLASS_NAME, 'store_name_map')
    for store_element in store_elements:
        store_name = store_element.find_element(By.TAG_NAME, 'i').text
        address = store_element.find_element(By.TAG_NAME, 'span').text.strip()

        store_info = {
            'store_name': store_name,
            'address': address
        }
        store_info_list.append(store_info)

    # "다음" 버튼 클릭
    if i == 5 or i == 10:
        try:
            next_button = driver.find_element(By.XPATH, '//a[contains(text(), "다음")]')
            next_button.click()
            WebDriverWait(driver, 10).until(EC.staleness_of(store_elements[0]))
        except NoSuchElementException:
            break

# 결과 출력
for store_info in store_info_list:
    print('Store Name:', store_info['store_name'])
    print('Address:', store_info['address'])
    print('---')

# WebDriver 종료
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

# Chrome WebDriver 경로 설정

# WebDriver 객체 생성
driver = webdriver.Chrome()
driver.implicitly_wait(3)

url = 'https://banapresso.com/store'
driver.get(url)

# 가게 정보를 담을 리스트
store_info_list = []

# 1부터 11까지의 버튼 클릭
for i in range(1, 12):
    # 버튼 클릭
    button_xpath = f'//li/a[contains(text(), "{i}")]'
    button = driver.find_element(By.XPATH, button_xpath)
    button.click()

    # 가게 정보 파싱
    store_elements = driver.find_elements(By.CLASS_NAME, 'store_name_map')
    for store_element in store_elements:
        store_name = store_element.find_element(By.TAG_NAME, 'i').text
        address = store_element.find_element(By.TAG_NAME, 'span').text.strip()

        store_info = {
            'store_name': store_name,
            'address': address
        }
        store_info_list.append(store_info)

    # "다음" 버튼 클릭
    if i == 5 or i == 10:
        try:
            next_button = driver.find_element(By.XPATH, '//a[contains(text(), "다음")]')
            next_button.click()
            WebDriverWait(driver, 10).until(EC.staleness_of(store_elements[0]))
        except NoSuchElementException:
            break

# 결과 출력
for store_info in store_info_list:
    print('Store Name:', store_info['store_name'])
    print('Address:', store_info['address'])
    print('---')

# WebDriver 종료
driver.quit()
```

```python
import os
import pandas as pd

# 폴더 경로
folder_path = './바나프레소'

# 폴더 생성
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

# 결과를 DataFrame으로 변환
df = pd.DataFrame(store_info_list)

# 엑셀 파일 경로
excel_file_path = os.path.join(folder_path, '바나프레소_가게정보.xlsx')

# 엑셀 파일로 저장
df.to_excel(excel_file_path, index=False)
print("가게 정보가 엑셀 파일로 저장되었습니다.")
```
