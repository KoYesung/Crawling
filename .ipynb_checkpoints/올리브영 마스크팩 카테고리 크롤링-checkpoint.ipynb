```python
import chromedriver_autoinstaller
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from urllib.request import Request, urlopen
import os
import requests
```

```python
try:
    driver = webdriver.Chrome()
    driver.implicitly_wait(3)

    url = 'https://www.oliveyoung.co.kr/store/display/getMCategoryList.do?dispCatNo=100000100090001&fltDispCatNo=&prdSort=01&pageIdx=1&rowsPerPage=24&searchTypeSort=btn_thumb&plusButtonFlag=N&isLoginCnt=0&aShowCnt=0&bShowCnt=0&cShowCnt=0&trackingCd=Cat100000100090001_Small'
    driver.get(url)
    time.sleep(3)
    image_urls = []

    for j in range(0,20):
        if j% 10 !=0:
            page_xpath = f'/html/body/div[3]/div[8]/div[2]/a[{j%10}]'
            if j>=10:
                page_xpath = f'/html/body/div[3]/div[8]/div[2]/a[{(j%10)+1}]'

            driver.find_element(By.XPATH, page_xpath).click()
            time.sleep(1) 
        for i in range(2,8):
            image_area_xpath = f'/html/body/div[3]/div[8]/div[1]/ul[{i}]'
            image_area = driver.find_element(By.XPATH, image_area_xpath)
            image_elements = image_area.find_elements(By.TAG_NAME, 'img')  # img태그를 전체 다 찾음
            # print(image_elements)
            for image_element in image_elements:
                image_url = image_element.get_attribute('src')
                print(image_url)
                image_urls.append(image_url)
                

        if j% 10 == 9:
            page_xpath = '/html/body/div[3]/div[8]/div[2]/a[10]'
            driver.find_element(By.XPATH, page_xpath).click()
            time.sleep(1) 

            
except:
    print('더 이상 제품이 없습니다!')
```

```python
url_splited = []
for url in image_urls:
    url2 =url.split('?')[0]
    url_splited.append(url2)
print(url_splited)
```

```python
import MySQLdb
import glob
import os

db = MySQLdb.connect(host='localhost', user='root', password='1234', db='kdt')
cur = db.cursor()

save_folder = "./마스크팩"
if not os.path.exists(save_folder):
    os.makedirs(save_folder)
    
# 이미지 다운로드 및 저장
for link in url_splited:
    response = requests.get(link)
    filename = link.split("/")[-1]
    filepath = os.path.join(save_folder, filename)  # 파일 경로 생성
    with open(filepath, "wb") as f:
        f.write(response.content)
        print(f"이미지 저장: {filename}")
    sql = "insert into OY_maskpack (img_name, img_path) values(%s, %s)"
    img_paths = os.path.join(save_folder, filename)  # 이미지 경로 생성
    data = [(filename, img_paths)]
    print(img_paths)
    cur.executemany(sql, data)  # data를 넣은 sql을 실행함
db.commit()  # 반영
```
